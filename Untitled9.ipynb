```python
import re

def decontracted(phrase):
    phrase = re.sub(r"\'t", "not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

for i in range(len(df)):
    df['Tweet'][i]=decontracted(df['Tweet'][i])
```

```python
list1=[]
list2=[]
list3=[]
list4=[]
list5=[]


for i in range(len(df)):
    if df.Label[i]=="Highly_Positive":
        list1.append(df['Tweet'][i].split())
    elif df.Label[i]=="Weakly_Positive":
        list2.append(df['Tweet'][i].split())
    elif df.Label[i]=="Neutral":
        list3.append(df['Tweet'][i].split())
    elif df.Label[i]=="Weakly_Negative":
        list4.append(df['Tweet'][i].split())
    else:
        list5.append(df['Tweet'][i].split())
```

# Highly Positive

```python
from collections import Counter
import itertools
all_words_no_urls1 = list(itertools.chain(*list1))

# Create counter
filtered_sentence1 = [w for w in all_words_no_urls1 if not w.lower() in stop_words]
counts_no_urls1 = Counter(filtered_sentence1)
counts_no_urls1.most_common(15)
```

```python
print(counts_no_urls1)
```

```python
clean_tweets_no_urls1 = pd.DataFrame(counts_no_urls1.most_common(15),
                             columns=['words', 'count'])

clean_tweets_no_urls1
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
clean_tweets_no_urls1.sort_values(by='count').plot.barh(x='words',
                      y='count',
                      ax=ax,
                      color="purple")

ax.set_title("Common Words Found in Tweets (Including All Words)")

plt.show()
```
